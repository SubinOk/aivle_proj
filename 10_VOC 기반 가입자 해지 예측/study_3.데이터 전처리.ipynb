- __주변값을 활용하여 결측치 채우기__<br>
fillna()의 method 파라미터를 사용하면, 사용자가 지정하는 일괄적인 값이 아닌 주변값으로 채울 수 있습니다.


```python
# 뒤에 있는 data를 사용해서 결측치를 처리하는 방법
df=df_fix.copy()
df=df.fillna(method='backfill')
df.head()
```



#### __<span style="color:RED">주의!!</span>__

#### __③ 결측치 제거하기__
결측치를 채우기로 처리하다보면 경우에 따라 값의 왜곡이 크게 발생하는 발생 하는 경우가 있습니다.    
이때는 제거하기 기법을 사용하여 결측치를 처리하면 데이터의 정합성을 보존 할 수 있습니다.

<br>

- __listwise 방식으로 제거하기__<br>
record의 항목 중 1개의 값이라도 NA이면 해당 데이터 행 전체 제거

```python
# listwise 방식으로 제거 하기
df=df_fix.copy()
df=df.dropna()
df.info()
```

- __pairwise 방식으로 제거하기__<br> 
모든 항목이 NA인 데이터 행만 제거

```python
# pairwise 방식으로 제거하기
df=df_fix.copy()
df=df.dropna(how='all')
df.info()
```

- __임계치를 설정해서 제거하기__<br>
NA의 갯수에 따라 결측데이터를 판단하고자 한다면, thresh 파라미터를 사용하면 됩니다.    


```python
# 임계치를 설정해서 제거하기
df=df_fix.copy()
df=df.dropna(thresh=14)
df.info()
```

- __특정 열에 있는 NA만 참고하여 결측치를 제거__<br>
Subset 파라미터를 사용하면 됩니다.

```python
# 특정열 안에 결측치가 있을 경우 삭제하기|
df=df_fix.copy()
df=df.dropna(subset=['cust_clas_itg_cd'])
df.info()
```






#### <font color='blue'>__[참고]__</font>
Q1와 Q3에서 IQR 거리의 1.5배가 넘어가는 값을 Outlier 라고 합니다.    
이 값들은 이상치로써 일반적으로 제거 또는 변경하여 데이터를 분석,학습 합니다.    
단, 이상치 분석시에는 제거하지 않습니다.    

<img src="https://wikidocs.net/images/page/166860/boxplot01.png
" width="600" >

[출처] https://wikidocs.net/166860

Median(중앙값) : 중앙값의 50% 위치<br>
Mean(평균값) : 평균값<br>
Box(박스) : Q1(25%) ~ Q3(75%) 까지 값들, IQR(Interquartile Range)<br>
Whiskers(수염) : Q1, Q3 로 부터 IQR 거리의 1.5배 내에서 가장 멀리 떨어진 데이터 점까지 이어져 있는 구역<br>
Outlier(이상치) : Whiskers 보다 바깥 쪽에 데이터가 존재하면, 통계적으로 이상치로 분류<br>

- __차트를 그려서 이상치 확인하기__<br>

이상치는 boxplot을 그려서 확인이 가능합니다. 


```python
plt.figure(figsize=(16,6))
sns.boxplot(df['age_itg_cd'])
plt.show()
```

- __이상치 제거하기__

```python
#Q1, Q3 구하기
q1 = df['age_itg_cd'].quantile(0.25)
q3 = df['age_itg_cd'].quantile(0.75)

# 1.5 * IQR(Q3 - Q1)
iqr = 1.5 * (q3 - q1) 

# 이상치 제거하기
df_data=df[(df['age_itg_cd'] < (q3 + iqr)) & (df['age_itg_cd'] > (q1 - iqr))]

df_data.describe()
```

- __이상치 변경하기__

```python
# 이상치 대체값 설정하기
age_min = q1 - iqr
age_max = q3 + iqr

# 이상치를 변경
df[(df['age_itg_cd'] > age_max)]= age_max
df[(df['age_itg_cd'] < age_min)]= age_min

df.describe()
```

```python
# 여기에 입력하세요.
cols = list(df)
for c in cols:
    if df[c].dtypes != object:
        df[c].fillna(df[c].median(), inplace=True)
```

즉, 연속형 변수를 범주형 변수로 만드는 방법이라고 보시면 됩니다.

 
#### __②  Encoding__
기계 학습의 주요 문제 중 하나는 많은 알고리즘이 범주 형 데이터를 입력값으로 수용하지 않는다는 점인데, 이를 Encoding을 통해 해결 할 수 있습니다.


- __Label Encoding__ : 문자를 숫자로 변환<br>
<img src="https://miro.medium.com/max/640/1*QQe-4476Oy3_dI1vhb3dDg.png" width="400" ><br>
문자형 카테고리를 숫자형 카테고리 값으로 변환합니다.<br>
하지만 레이블 인코딩이 일괄적으로 숫자 값으로 변환이 되면서 몇몇 ML알고리즘에는 이를 적용할 경우 예측 성능이 떨어지는 경우가 발생할 수 있습니다.<br>
<font color="red">__숫자로 되어 있어 잘못하면 가중치로 인식하여 값에 왜곡이 생기게 됩니다.__</font><br>
이러한 특성 때문에 레이블 인코딩은 선형 회귀에는 적용하지 않습니다.<br>



```python
# 라이브러리 임포트
from sklearn.preprocessing import LabelEncoder
```
