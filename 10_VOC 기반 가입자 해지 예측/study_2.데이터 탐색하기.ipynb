#### __<font color='red'>[Q]</font> 'df'에서 컬럼 별 '_'가 50% 이상 차지하는 컬럼들을 찾으세요.__


```python
# 여기에 입력하세요.
cols = list(df)
dummies = []

for c in cols:
    vc = df[c].value_counts(normalize=True)
    try:
        tmp = vc['_']
        if tmp >= 0.5:
            dummies.append(c)
    except:
        continue
```
