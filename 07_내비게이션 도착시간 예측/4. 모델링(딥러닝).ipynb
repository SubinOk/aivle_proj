* dense와 dense_1의 activation은 'relu'으로 지정해주세요.
* 모델 compile 시 loss는 'mse', optimizer는 'adam', metrics는 'mae'와 'mse'로 지정해주세요.
* 모델 학습 시에는 콜백 함수 두 가지를 적용해주세요.(EarlyStopping : val_loss기준 patience 10, ModelCheckpoint : val_loss기준 save_weights_only, save_best_only)

![model_architecture.png](./model/model_architecture.png)

```python
# 아래에 실습코드를 작성하세요.
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
inputs = Input(shape=(108, ))
x = inputs

x = Dense(64, activation='relu')(x)
x = Dense(64, activation='relu')(x)

outputs = Dense(1)(x)

model = keras.models.Model(inputs, outputs)
model.compile(optimizer='adam', loss='mse', metrics=['mae', 'mse'])

model.summary()
```

```python
es = EarlyStopping(monitor='val_loss', min_delta=0, patience=10)
chk = ModelCheckpoint(filepath='./model/cp.ckpt', monitor='val_loss', save_weights_only=True, save_best_only=True)
```
