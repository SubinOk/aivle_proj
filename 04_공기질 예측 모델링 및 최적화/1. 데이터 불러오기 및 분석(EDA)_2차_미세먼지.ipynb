```python
air_21['측정일시'] = pd.to_datetime(air_21['측정일시'], format='%Y%m%d%H')
air_22['측정일시'] = pd.to_datetime(air_22['측정일시'], format='%Y%m%d%H')
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot_date(weather_21_s['일시'], weather_21_s['기온(°C)'], 'r-')
ax2.plot_date(weather_21_s['일시'], air_tmp['PM10'], 'g-')

ax1.set_xlabel('일시')
ax1.set_ylabel('기온')
ax2.set_ylabel('PM10')

plt.show()
```
