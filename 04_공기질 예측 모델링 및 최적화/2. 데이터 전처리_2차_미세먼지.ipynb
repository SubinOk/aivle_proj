# (실습준비) 데이터 불러오기부터

```python
import numpy as np
import pandas as pd
```

```python
# 학습 데이터 로딩
air_21 = pd.read_csv("air_2021.csv", sep=',', index_col = 0, encoding = 'utf-8' )
air_22 = pd.read_csv("air_2022.csv", sep=',', index_col = 0, encoding = 'utf-8' )
weather_21 = pd.read_csv("weather_2021.csv", sep = ',', encoding='cp949')
weather_22 = pd.read_csv("weather_2022.csv", sep = ',', encoding='cp949')
```

# 1. 이상치/결측치 처리

#### [실습문제1] 미세먼지 데이터 결측치 처리
* 미세먼지 데이터의 결측치를 확인하고 결측치가 존재한다면 처리해주세요.


* 앞서 데이터 불러오기 및 분석 단계에서 확인한 불필요한 변수에 대해 제거 해 주세요. 

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
col = ['지역', '망', '측정소코드', '측정소명', '주소']

air_21 = air_21.drop(col, axis=1)
air_22 = air_22.drop(col, axis=1)
```

```python
air_21.isna().sum() # 전체 8760 행
```

```python
air_22.isna().sum() # 전체 2159 행
```

- NaN 값 처리

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
col_name = list(air_21)
```

```python
imputer = IterativeImputer()
air_21 = imputer.fit_transform(air_21)
air_22 = imputer.fit_transform(air_22) # test 데이터이므로 따로 설정
```

```python
air_21= pd.DataFrame(air_21)
air_22 = pd.DataFrame(air_22)
```

```python
air_21.columns = col_name
air_22.columns = col_name
```

```python
air_21.head()
```

#### [실습문제2] 날씨 데이터 결측치 처리
* 날씨 데이터의 결측치를 확인하고 결측치가 존재한다면 처리해주세요.

* '강우량'을 어떻게 해야할까요?

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
col = ['지점', '지점명', '기온 QC플래그', '강수량 QC플래그', '풍속 QC플래그', '풍향 QC플래그', '습도 QC플래그', '현지기압 QC플래그', '해면기압 QC플래그', '3시간신적설(cm)', '지면상태(지면상태코드)', '지면온도 QC플래그', '현상번호(국내식)', '일조 QC플래그', '일사 QC플래그', '운형(운형약어)']

weather_21 = weather_21.drop(col, axis=1)
weather_22 = weather_22.drop(col, axis=1)
```

```python
weather_21.isna().sum() # 전체 8759행
```

```python
weather_22.isna().sum() # 전체 2159행
```

```python
na_cols = ['강수량(mm)', '적설(cm)']
weather_21 = weather_21.drop(na_cols, axis=1)
weather_22 = weather_22.drop(na_cols, axis=1)
```

- NaN 값 처리

```python
col_name = list(weather_21)
```

```python
day_21 = weather_21.loc[:, ['일시']]
day_22 = weather_22.loc[:, ['일시']]
```

```python
tmp_21 = weather_21.drop('일시', axis=1)
tmp_22 = weather_22.drop('일시', axis=1)
```

```python
imputer = IterativeImputer()
tmp_21 = imputer.fit_transform(tmp_21)
tmp_22 = imputer.fit_transform(tmp_22) # test 데이터이므로 따로 설정
```

```python
tmp_21 = pd.DataFrame(tmp_21)
tmp_22 = pd.DataFrame(tmp_22)
```

```python
weather_21 = pd.concat([day_21, tmp_21], axis=1)
weather_22 = pd.concat([day_22, tmp_22], axis=1)
```

```python
weather_21.columns = col_name
weather_22.columns = col_name
```

```python
weather_22.head()
```

#### [실습문제3] air_21, air_22 의 '측정일시'를 활용하여 'time'변수 생성하기


* air_21, air_22 의 '측정일시'를 활용하여 'time'변수 생성


* time 변수를 datatime으로 데이터 타입 변경하기

```python
# 1일 24시 = 2일 0시로 데이터가 설정되어 있어서 24 -> 0으로 바꾸면 시간대 오류 발생
air_21['time'] = air_21['측정일시'] - 1
air_22['time'] = air_22['측정일시'] - 1

air_21['time'] = pd.to_datetime(air_21['time'],format='%Y%m%d%H')
air_22['time'] = pd.to_datetime(air_22['time'],format='%Y%m%d%H')
```

```python
import datetime

air_21['time']= air_21['time'] + datetime.timedelta(hours=1)
air_22['time']= air_22['time'] + datetime.timedelta(hours=1)
```

```python
air_21.drop('측정일시', axis=1, inplace=True)
air_22.drop('측정일시', axis=1, inplace=True)
```

#### [실습문제4] weather_21, weather_22 의 '일시'를 활용하여 'time'변수 생성하기


* weather_21, weather_22 의 '측정일시'를 활용하여 'time'변수 생성


* time 변수를 datatime으로 데이터 타입 변경하기

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
weather_21['time'] = pd.to_datetime(weather_21['일시'])
weather_22['time'] = pd.to_datetime(weather_22['일시'])
```

```python
weather_21.drop('일시', axis=1, inplace=True)
weather_22.drop('일시', axis=1, inplace=True)
```

# 2. 데이터 합치기
* 미세먼지 데이터와 날씨 데이터를 'time' 기준으로 합쳐보겠습니다

#### [실습문제5] 데이터 합치기
* 'df_21','df_22'에 21년도와 22년도의 미세먼지, 날씨 데이터를 합쳐 넣도록 하겠습니다.

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다
df_21 = pd.merge(air_21, weather_21, how='inner', on='time')
df_22 = pd.merge(air_22, weather_22, how='inner', on='time')
```

```python
df_21 = df_21.drop('time', axis=1)
df_22 = df_22.drop('time', axis=1)
```

```python
df_21.head()
```

#### [실습문제6] 각종 인자 추가 및 사용하지 않을 변수 제거

* 새로운 인자를 추가 하세요.


* 머신러닝에 사용하지 않을 변수들을 제거해줍니다.

```python
df_21.corr()
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다
p_less_1p = ['O3', '풍속(m/s)', '풍향(16방위)', '습도(%)', '현지기압(hPa)', '해면기압(hPa)', '일조(hr)', '일사(MJ/m2)', '중하층운량(10분위)', '최저운고(100m )']
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다
df_21 = df_21.drop(p_less_1p, axis=1)
df_22 = df_22.drop(p_less_1p, axis=1)
```

#### [실습문제7] 미세먼지 농도 다음시점 데이터 생성

* t+1 시점 데이터 생성

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다
df_21['PM10_lag1'] = df_21['PM10'].shift(-1)
df_22['PM10_lag1'] = df_22['PM10'].shift(-1)
```

```python
df_21.head()
```

```python
l_21 = df_21.shape[0] - 1
l_22 = df_22.shape[0] - 1
```

```python
df_21 = df_21.drop(l_21, axis=0)
df_22 = df_22.drop(l_22, axis=0)
```

# 3. train / test split

#### [실습문제7] 데이터셋 나누기 

* 데이터셋을 train(21년도 데이터), test(22년도 데이터)로 나눠 csv로 저장하도록 하겠습니다.


* train_x / train_y / test_x / test_y 로 나누세요.


* train_x.csv / train_y.csv / test_x.csv / test_y.csv 로 저장하세요.

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
train_x = df_21.drop('PM10_lag1', axis=1)
train_y = df_21['PM10_lag1']

test_x = df_22.drop('PM10_lag1', axis=1)
test_y = df_22['PM10_lag1']
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
train_x.to_csv('train_x.csv', index=False)
train_y.to_csv('train_y.csv', index=False)

test_x.to_csv('test_x.csv', index=False)
test_y.to_csv('test_y.csv', index=False)
```
